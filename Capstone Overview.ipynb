# Collaborative Filtering Algorithm

This algorithm uses past consumer behavior to recommend what they will like in the future.  It matches consumers to items that consumers like them *also* like, but they have not consumed/watched/bought/reviewed yet.   So if consumer A likes restaurants 1,2,3,4  and consumer B likes restaurants 2,3,4,5, the algorithm will predict that consumer A will like restaurant 5, and that consumer B will like restaurant 1.  There are two different routes I can go with this.

> One of the most popular examples of this is Amazon's product recommendations.
> There is also a content-based algorithm, which recommends an item based on similarity to an item the user has already liked.  This works well with items that are easily determinable in terms of context and properties.   This may actually work really well for restaurants!   Yelp restaurants could be easy to sort based on certain properties using NLP.

## 1. User-User Collaborative Filtering

This form of filtering recommends to a user a product that a look-alike consumer has chosen in the past.  In this particular case, it would be consumer A being recommended restaurant 5 because their look-alike, consumer B, reviewed in highly in the past.  This algorithm is very effective but also computationally expensive,  as it needs to compute every consumer pair.


## 2. Item-Item Collaborative Filtering

This is quite similar to before, but instead of consumer look-alikes, we're now looking for product look-alikes.  In this case, restaurants.  I would have to create a restaurant look-alike matrix, but then the recommendations would be based on what items are like what the consumer has already positively engaged with (in this case reviewed highly).  This is much less computationally expensive.  I am working with a fixed dataset, so there is a limit to how many restaurants are in this matrix!  So the matrix itself will be fixed.
